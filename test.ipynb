```python
import pandas as pd

# Assuming your DataFrame is called df
# Filter for the year 2025
df_2025 = df[df['Date'].dt.year == 2024]

# Group by month and sum Quantity
monthly_sum = df_2025.groupby(df_2025['Date'].dt.month)['Quantity'].sum()

# Optional: rename the index to month names
monthly_sum.index = monthly_sum.index.map(lambda x: pd.to_datetime(str(x), format='%m').strftime('%B'))

print(monthly_sum)
```
